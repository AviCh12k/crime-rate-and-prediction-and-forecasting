```python
api = tweepy.API(auth)
#made a cursor
c = tweepy.Cursor(api.search, q='cyber crime india')
c.pages(5000) # you can change it make get tweets

#Lets save the selected part 
# x = []
for tweet in c.items():
    text.append(tweet._json['text'])
    time.append(tweet._json['created_at'])
    location.append(tweet._json['location'])
    category.append('cyber crime')
```
